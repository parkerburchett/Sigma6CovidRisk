```python
combined_filter = df['medcond_yn'].str.contains('Yes|No') & df['death_yn'].str.contains('Yes|No') & df['cdc_case_earliest_dt '].notna()


combined_filter.value_counts()
```

```python
rel_df = df[combined_filter]
```

```python
chunks = pd.read_csv('COVID-19_Case_Surveillance_Public_Use_Data (1).csv', chunksize=1_000_000)
df_to_save = pd.DataFrame(columns=rel_df.columns) 
for c in chunks:
    combined_filter = c['medcond_yn'].str.contains('Yes|No') & c['death_yn'].str.contains('Yes|No') & c['cdc_case_earliest_dt '].notna() & c['sex'].notna() & c['age_group'].notna() &  c['race_ethnicity_combined'].notna() 
    df_to_save = df_to_save.append(c[combined_filter])
    print(df_to_save.shape)
    
df['month_year'] = pd.to_datetime(df['cdc_case_earliest_dt ']).dt.to_period('M')
df_to_save.to_csv('cleaned_surveillance_data.csv')
```
